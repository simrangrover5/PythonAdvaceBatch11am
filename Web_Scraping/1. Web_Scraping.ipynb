```python
for i in soup.select("li"):  #<li>Hello</li>
    print(i.text)
```

```python
for ele in soup.select("p"):
    print(ele.text)
```

```python
for i in soup.select("a"):
    print(i.text)
```

```python
# we want href attribute of a tag
```

```python
for i in soup.select("a"):
    print(i['href'])
```

```python
https://www.scrapinghub.com/data-services/
```

```python
domain = "https://www.scrapinghub.com"
for i in soup.select("a"):
    if i['href'].startswith("/"):
        href = domain + i['href']
    else:
        href = i['href']
    print(href)
```

```python
for i in soup.select("img"):
    print(domain + i['src'])
```
