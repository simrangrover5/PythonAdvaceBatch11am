#### WITH 

    Context Manager
    It manages the resource and if you open or try to access some resources but you did not close it
    In that case use context manager which will automatically close the resourse

```python
for i in range(5000):
    fp = open("test.py")
```

```python
with open("test.py","w") as fp:
    print(fp)
    fp.write("Hello this is test file for with")
```

```python
fp.write("SIMRAN GROVER")   #bcz outside the indent the with will automatically close the file so you cannot access the file
#outside the indent
```

```python
with open("one.py","w+") as fp:
    fp.write("a = input()\nb = input()")  #cursor will react to the end
    fp.seek(0)  #so to bring the cursor again on top 
    print(fp.read())
```
