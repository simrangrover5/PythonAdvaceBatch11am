```python
i = 1
while i<=4:
    j = 1
    while j<=4:
        print(i,j)
        j += 1
        print("*"*80)
    print("#"*80)
    i += 1
#i = 1, j = 1 --> 1,1,j +=1 --> 2
#i = 1, j = 2 , j += 1 --> 3
#i = 1, j = 3 , j += 1 --> 4
#i = 1, j = 4, j += 1 --> 5
# nested loop condition becomes false
```

```python
i = 0
while i<2:
    print("hi")
    j = 0
    while j<2:
        print("hello world")
        j += 1
    i += 1
#1. 0<2 --> True --> hi, j = 0, 0<2
#        hello world, j += 1 --> 0 + 1 --> 1
#   1<2 --> Hello world , j += 1 --> 1 + 1 --> 2
#   2<2 --> False (exit inner loop)
#   i += 1 --> 0 + 1 --> 1
#2. 1<2 --> True --> hi, j = 0, 
#   0<2 --> hello world --> j += 1 --> 0 + 1 --> 1
#   1<2 --> hello world --> j += 1 --> 1 + 1 --> 2
#   2<2 --> False exit nested the loop
#   i += 1--> 1 + 1 = 2
#3. 2<2  --> False --> Exit main loop
```

```python
n = int(input("\n Enter any number : "))
row = 1   # formula --> n - (row-1)
while row<=n:
    col = 1
    while col<=(n-(row-1)):
        print("*",end=" ")
        col += 1
    print()
    row += 1
```

```python
n = int(input("\n Enter any number : "))
row = 0  # formula --> n - (row)
while row<n:
    col = 1
    while col<=(n-(row)):
        print("*",end=" ")
        col += 1
    print()
    row += 1
#5 - 0 = 5
#5 - 1 = 4
#5 - 2 = 3
#5 - 3 = 2
#5 - 4 = 1
```

```python
n = int(input("\n Enter any number : "))
row = n 
while row>0:
    col = 1
    while col<=row:
        print("*",end=" ")
        col += 1
    print()
    row -= 1
# row = 5, col= 1,2,3,4,5
# row -= 1 --> 4, col = 1,2,3,4
# row -= 1 --> 3, col = 1,2,3
# row -= 1 --> 2, col = 1,2
# row -= 1 --> 1 ,col = 1
```
