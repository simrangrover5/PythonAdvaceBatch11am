```python
from time import sleep
from getpass import getpass
while True:
    print("\n1. Login\n2. Signup\n3. Forgot Password\n4. Exit")
    ch = int(input("\n Choose any option from above 1/2/3/4 : "))
    if ch == 1:
        print("LOGGING IN........")
        accno = input("\n Enter your account number : ")
        if accno in data.keys():
            pswd = getpass("\n Enter your password : ")
            if pswd == data[accno][1]:
                print("\n WELCOME.....")
                while True:
                    print("\n1. Credit\n2. Debit\n3. Show Details\n4. Change Password\n5. Logout")
                    ch = int(input("\n Enter any option from above : "))
                    if ch == 1:
                        pass
                    elif ch == 2:
                        pass
                    elif ch == 3:
                        d = data[accno]
                        print("DETAILS".center(80,'*'))
                        print(f"Name : {d[0]}".center(80))
                        print(f"Password : {d[1]}".center(80))
                        print(f"Balance : {d[2]}".center(80))
                        print(f"Security Answer : {d[3]}".center(80))
                    elif ch == 4:
                        old_passwd = getpass("\n Enter your old password")
                        if old_passwd == data[accno][1]:
                            newpass = getpass("\n Enter your new password : ")
                            data[accno][1] = newpass
                            print("\n Successfully Changed The Password....")
                        else:
                            print("\n Old password does not matched.....")
                    elif ch == 5:
                        print("Logging Out.......")
                        sleep(2)
                        print("\n Thankyou for login.....")
                        break
                    else:
                        print("\n Invalid Option......")
            else:
                print("\n Password Does not match try again")
        else:
            print("\n Invalid Account number.....")
    elif ch == 2:
        print("\n WELCOME TO SIGNUP PAGE")
        name = input("\n Enter your name : ")
        passwd = getpass("\n Enter your password : ")
        if len(passwd) >= 8:
            l = 0
            u = 0
            s = 0
            n = 0
            for i in passwd:
                if i in "abcdefghijklmnopqrstuvwxyz":
                    l += 1
                elif i.isupper():
                    u += 1
                elif i in "0123456789":
                    n += 1
                elif i in ['#','$','%','&','@',' ','.','!','*']:
                    s += 1
            if l>=1 and u>=1 and s>=1 and n>=1:
                bal = int(input("\n ENter your initial balance : "))
                if bal >= 10000:
                    date = input("\n Security Question (Enter your dob in ddmm format) : ")
                    d = {("01","03","05","07","08","10","12"):31,("02",):29,("04","06","09","11"):30}
                    m = date[2:]
                    d1 = date[:2]
                    if 1<=int(m)<=12:
                        for k in d.keys():
                            if m in k:
                                v = d[k]
                                if 1<=int(d1)<=v:
                                    break
                        else:
                            print("\n Invalid Date Please Enter correct date")
                            continue

                    else:
                        print("\n Invalid Date...Please Enter correct date")
                        continue
                    newacc = int(list(data.keys())[-1]) + 1
                    data[str(newacc)] = [name,passwd,bal,date]
                    print(f"\n Your account number is {newacc} and thanks for being customer of your bank....")
                else:
                    print("\n Your initial balance is insufficient....It must be 10000 atleast")
            else:
                print("\n Password is invalid")
        else:
            print("\n Password Must be 8 characters long")
    elif ch == 3:
        pass
    elif ch == 4:
        print("\n ThankYou for Using our services......")
        break
    else:
        print("\n Invalid Choice.......")
```

```python
def password_validate(passwd):
    if type(passwd) == str:
        if len(passwd) >= 8:
            l = 0
            u = 0
            s = 0
            n = 0
            for i in passwd:
                if i in "abcdefghijklmnopqrstuvwxyz":
                    l += 1
                elif i.isupper():
                    u += 1
                elif i in "0123456789":
                    n += 1
                elif i in ['#','$','%','&','@',' ','.','!','*']:
                    s += 1
            if l>=1 and u>=1 and s>=1 and n>=1:
                return True
            else:
                return False
        else:
            return False
    else:
        return False
        
```
