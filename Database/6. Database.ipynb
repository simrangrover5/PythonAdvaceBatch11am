1. Which languages do people speak having the country name starting from A?

```python
%%sql

select language, name from countrylanguage inner join country on countrylanguage.countrycode = country.code where country.name like "a%";
```

2. Find the lifeexpectancy of people speaking offical english in there country.

```python
%%sql

select Name, LifeExpectancy from country inner join countrylanguage on country.code= countrylanguage.countrycode where language = "english" and isofficial = "T";
```

3. Find data from countrylanguage for these codes ABW, AFG and AGO

```python
%%sql

Select * from countrylanguage where countrycode in ("AFG","AGO","ABW")
```

4. Find total population speaking dutch language in their country as official language.

```python
%%sql

select sum(population) from country inner join countrylanguage on country.code = countrylanguage.countrycode where language="dutch";
```
